```python
#28. maketrans- method is used to create a translation table.  Doubt

#This table is mainly used with the .translate() method to replace or remove characters in a string.

# str.maketrans(x, y, z)
# x → characters to replace
# y → replacement characters
# z → characters to remove (optional)

# x and y must have the same length.

table1 = str.maketrans("aup", "123") #replace
a = "Rupak Sharma"
print(a.translate(table1))

table2 = str.maketrans("a", "1", "o") #replace and remove
b = "a23456674332aaoo"
print(b.translate(table2))
```
